```python
# -------------------------
# Netflix Content Trends Analysis
# -------------------------
# ✅ Tested on Pandas 2.x (Google Colab)
# ✅ Implements EXACT Problem Statement:
#    1. Movies vs TV Shows over years
#    2. Genre popularity trends
#    3. Country-wise contributions
# -------------------------

# (1) Install required
!pip install plotly --quiet

# (2) Imports
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from google.colab import files

sns.set(style="whitegrid")
plt.rcParams['figure.dpi'] = 120

# (3) Upload dataset
print("▶ Please upload your Netflix CSV file (e.g., 'Netflix Dataset.csv').")
uploaded = files.upload()
if uploaded:
    csv_file = list(uploaded.keys())[0]
else:
    csv_file = "Netflix Dataset.csv"
print("Using file:", csv_file)

# (4) Read CSV safely
try:
    df = pd.read_csv(csv_file)
except Exception as e:
    print("utf-8 failed, retrying with latin1:", e)
    df = pd.read_csv(csv_file, encoding='latin1')

# (5) Normalize column names
df.columns = (
    df.columns
    .astype(str)
    .str.strip()
    .str.lower()
    .str.replace(r'\s+', '_', regex=True)
    .str.replace('-', '_', regex=False)
)

print("\n--- Initial Info ---")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())
display(df.head(3))

# (6) Clean duplicates & strings
df = df.drop_duplicates().reset_index(drop=True)
for c in df.select_dtypes(include=['object']).columns:
    df[c] = df[c].astype(str).str.strip().replace({'nan': np.nan})

# (7) Handle dates & years
if 'date_added' in df.columns:
    df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
    df['date_added_year'] = df['date_added'].dt.year
else:
    df['date_added_year'] = pd.NA

if 'release_year' in df.columns:
    df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').astype('Int64')
else:
    df['release_year'] = pd.NA

df['analysis_year'] = df['date_added_year'].fillna(df['release_year'])
df['analysis_year'] = pd.to_numeric(df['analysis_year'], errors='coerce')

print("\nAfter cleaning: shape =", df.shape)
display(df.head(2))

# -------------------------
# Detect important columns
# -------------------------
def find_col(*keywords):
    keys = [k.lower() for k in keywords]
    for c in df.columns:
        for k in keys:
            if k in c:
                return c
    return None

type_col = find_col('type')                  # 'Movie' or 'TV Show'
genre_col = find_col('listed_in', 'genre')
country_col = find_col('country')

print("\nDetected columns:")
print("type_col:", type_col)
print("genre_col:", genre_col)
print("country_col:", country_col)

os.makedirs("plots", exist_ok=True)

# -------------------------
# Objective 1: Movies vs TV Shows over years
# -------------------------
if type_col:
    df_type_year = (
        df.dropna(subset=['analysis_year', type_col])
          .groupby(['analysis_year', type_col])
          .size()
          .reset_index(name='count')
    )

    df_type_year['analysis_year'] = pd.to_numeric(df_type_year['analysis_year'], errors='coerce')
    df_type_year['count'] = pd.to_numeric(df_type_year['count'], errors='coerce')

    print("\n>>> Sample (Movies vs TV Shows by year):")
    display(df_type_year.head())

    pivot = df_type_year.pivot(index='analysis_year', columns=type_col, values='count').fillna(0).sort_index()
    pivot = pivot.apply(pd.to_numeric, errors='coerce').fillna(0)

    if not pivot.empty:
        ax = pivot.plot(kind='bar', stacked=True, figsize=(12,5))
        ax.set_xlabel("Year")
        ax.set_ylabel("Number of Titles")
        ax.set_title("Movies vs TV Shows — Titles by Year")
        plt.tight_layout()
        plt.savefig("plots/type_by_year_stacked.png")
        plt.show()
    else:
        print("⚠ No valid data available for plotting Movies vs TV Shows trend.")

    if not df_type_year.empty:
        fig = px.bar(df_type_year, x='analysis_year', y='count', color=type_col, barmode='stack',
                     title="Movies vs TV Shows over Years (interactive)")
        fig.update_layout(xaxis_title='Year', yaxis_title='Count')
        fig.show()
else:
    print("\n⚠ No 'type' column found — skipping Objective 1.")

# -------------------------
# Objective 2: Genre popularity trends
# -------------------------
if genre_col:
    df['__genres_list'] = df[genre_col].fillna('').astype(str).str.split(',').apply(lambda lst: [g.strip() for g in lst if g.strip()])
    genres_exploded = df.explode('__genres_list').rename(columns={'__genres_list':'genre'})

    top_genres = genres_exploded['genre'].value_counts().head(12)
    print("\n>>> Top Genres Overall:")
    display(top_genres)

    plt.figure(figsize=(10,5))
    top_genres.sort_values().plot(kind='barh')
    plt.title("Top Genres (Overall)")
    plt.xlabel("Count")
    plt.tight_layout()
    plt.savefig("plots/top_genres_overall.png")
    plt.show()

    topN = top_genres.index.tolist()
    trend = (genres_exploded.dropna(subset=['analysis_year'])
             .loc[genres_exploded['genre'].isin(topN)]
             .groupby(['analysis_year','genre']).size()
             .reset_index(name='count'))

    if not trend.empty:
        pivot = trend.pivot(index='analysis_year', columns='genre', values='count').fillna(0).sort_index()
        pivot = pivot.apply(pd.to_numeric, errors='coerce').fillna(0)

        if not pivot.empty:
            pivot.plot(kind='line', figsize=(12,6), marker='o')
            plt.title("Trend of Top Genres Over Years")
            plt.ylabel("Count")
            plt.xlabel("Year")
            plt.legend(title='Genre', bbox_to_anchor=(1.05,1), loc='upper left')
            plt.tight_layout()
            plt.savefig("plots/top_genres_trend.png")
            plt.show()

        fig = px.line(trend, x='analysis_year', y='count', color='genre', markers=True,
                      title="Top Genres Trend Over Years (interactive)")
        fig.show()
else:
    print("\n⚠ No genre column found — skipping Objective 2.")

# -------------------------
# Objective 3: Country-wise contributions
# -------------------------
if country_col:
    df['__country_list'] = df[country_col].fillna('').astype(str).str.split(',').apply(lambda lst: [c.strip() for c in lst if c.strip()])
    countries_exploded = df.explode('__country_list').rename(columns={'__country_list':'country_clean'})

    top_countries = countries_exploded['country_clean'].value_counts().head(12)
    print("\n>>> Top Countries by Titles:")
    display(top_countries)

    plt.figure(figsize=(10,5))
    top_countries.sort_values().plot(kind='barh')
    plt.title("Top Countries by Titles")
    plt.xlabel("Count")
    plt.tight_layout()
    plt.savefig("plots/top_countries.png")
    plt.show()

    topC = top_countries.index.tolist()
    ctrend = (countries_exploded.dropna(subset=['analysis_year'])
              .loc[countries_exploded['country_clean'].isin(topC)]
              .groupby(['analysis_year','country_clean']).size()
              .reset_index(name='count'))

    if not ctrend.empty:
        pivot = ctrend.pivot(index='analysis_year', columns='country_clean', values='count').fillna(0).sort_index()
        pivot = pivot.apply(pd.to_numeric, errors='coerce').fillna(0)

        if not pivot.empty:
            pivot.plot(kind='line', figsize=(12,6), marker='o')
            plt.title("Top Countries Contribution Trend")
            plt.ylabel("Count")
            plt.xlabel("Year")
            plt.legend(title='Country', bbox_to_anchor=(1.05,1), loc='upper left')
            plt.tight_layout()
            plt.savefig("plots/top_countries_trend.png")
            plt.show()

        fig = px.line(ctrend, x='analysis_year', y='count', color='country_clean', markers=True,
                      title="Top Countries Trend (interactive)")
        fig.show()
else:
    print("\n⚠ No country column found — skipping Objective 3.")

# -------------------------
# Save cleaned dataset
# -------------------------
cleaned_name = "netflix_cleaned.csv"
df.to_csv(cleaned_name, index=False)
print(f"\n✅ Cleaned dataset saved to: {cleaned_name}")
try:
    files.download(cleaned_name)
except:
    print("Download from Colab file panel if auto-download fails.")

# -------------------------
# Quick Summary
# -------------------------
print("\n===== Summary =====")
if type_col: print("\nType distribution:\n", df[type_col].value_counts())
if genre_col: print("\nTop genres:\n", genres_exploded['genre'].value_counts().head(10))
if country_col: print("\nTop countries:\n", countries_exploded['country_clean'].value_counts().head(10))

# -----------------------------
# Strategic Recommendations
# -----------------------------
print("\n===== Strategic Recommendations =====")
print("\n1. Strengthen high-demand genres (Drama, Comedy, Thriller)")
print("\n2. Invest in underrepresented genres (Documentary, Animation)")
print("\n3. Balance Movie vs TV Show releases")
print("\n4. Expand region-specific content (India, South Korea, Latin America)")
print("\n5. Global reach with local appeal (subtitles/dubbing)")
print("\n6. Data-driven content acquisition (engagement metrics)")
print("\n7. Experiment with innovative formats (interactive, short-form)")

```
